## Line Chart

```
import numpy as np
import matplotlib.pyplot as plt
```

```
x = np.linspace(0, 20, 1000)
```

```
y = np.sin(x) + 0.2 * x
```

```
plt.plot(x, y)
```

```
# if you are not in a notebook
plt.plot(x, y)
plt.show()
```

```
plt.plot(x, y)
plt.xlabel('input')
plt.ylabel('output')
plt.title('my plot');
```

## Scatterplot

```
X = np.random.randn(100, 2)
plt.scatter(X[:,0], X[:,1]);
```

```
X = np.random.randn(200, 2)
X[:50] += 3
Y = np.zeros(200)
Y[:50] = 1
plt.scatter(X[:,0], X[:,1], c=Y);
```
