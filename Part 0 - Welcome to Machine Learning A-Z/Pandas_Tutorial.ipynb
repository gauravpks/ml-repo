## The apply() function


```
# an in python for loops are slow
def date_to_year(row):
  return int(row['date'].split('-')[0]) # Grabs the 0th element

df.apply(date_to_year, axis=1) ## Axis is required else will work column wize
```

```
df['year'] = df.apply(date_to_year, axis=1)
```
